````python
metadata_raw = generate_metadata(text)

print("📦 Raw LLM Output:\n")
print(metadata_raw)
````

````python
import re
import json

# Extract JSON string from Markdown-style output
match = re.search(r"```(?:json)?\s*(\{.*?\})\s*```", metadata_raw, re.DOTALL)
if match:
    json_str = match.group(1)
    metadata_json = json.loads(json_str)
    display(JSON(metadata_json))
else:
    print("❌ Could not extract JSON from raw output.")
````

````python
import pandas as pd

# Flatten JSON for tabular display
flat_metadata = {k: str(v) for k, v in metadata_json.items()}
pd.DataFrame(flat_metadata.items(), columns=["Metadata Field", "Value"])
````

````python
from IPython.display import display, HTML

def display_metadata_html(metadata):
    html = "<table style='width:100%; border-collapse: collapse;'>"
    for key, value in metadata.items():
        html += f"<tr><td style='padding:8px; border:1px solid #ddd; font-weight:bold;'>{key}</td>"
        html += f"<td style='padding:8px; border:1px solid #ddd;'>{value}</td></tr>"
    html += "</table>"
    display(HTML(html))

display_metadata_html(metadata_json)
````

````python
from IPython.display import display, HTML

def collapsible_metadata_view(metadata):
    html = "<details><summary><b>Click to expand metadata</b></summary><pre style='font-family:monospace;'>"
    for key, value in metadata.items():
        html += f"{key}: {value}\n"
    html += "</pre></details>"
    display(HTML(html))

collapsible_metadata_view(metadata_json)
````

````python
with open("output_metadata.json", "w") as f:
    json.dump(metadata_json, f, indent=4)
print("✅ Metadata saved to output_metadata.json")
````

````python
def to_markdown(metadata):
    return "\n".join([f"- **{k}**: {v}" for k, v in metadata.items()])

print(to_markdown(metadata_json))
````
